### Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### **Data Acquisition & Inspection**

This will show the first 10 rows of the data and how it's structured. There are 150 rows & 5 columns. Columns sepal length, sepal width, petal length, and petal width are numerical. Column species is categorical. 

```python
# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

### **Initial Descriptive Statistics**

This shows the descriptive statistics of the data

```python
print(df.describe())
```

### **Initial Data Distribution for Numerical Columns**

This shows all the numerical columns as a histogram. The petal width histogram is displayed larger to see.

```python
# Inspect histogram by numerical column
df['petal_width'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

### Document Observations:
- The sepal length and width are normally distributed compared to the petal length and width
- The petal length & width have large values between 0 and 0.5

### **Initial Data Distribution for Categorical Columns**

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

### Document Observations: 
- There are 3 species of Iris flowers: Setosa, Versicolor, and Virginica
- They each have a count of 50

### **Initial Data Transformation and Feature Engineering**

Rename columns to look cleaner. Add more columns for more data.

```python
# Renaming a column
df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)
df.rename(columns={'sepal_width': 'Sepal Width'}, inplace=True)
df.rename(columns={'petal_length': 'Petal Length'}, inplace=True)
df.rename(columns={'petal_width': 'Petal Width'}, inplace=True)
df.rename(columns={'species': 'Species'}, inplace=True)

# Adding a new column
df['Sepal Area'] = df['Sepal Length'] * df['Sepal Width']
df['Petal Area'] = df['Petal Length'] * df['Petal Width']
```

### **Initial Visualizations**

```python
sns.pairplot(df, hue='Species')
plt.show()
```

### **Initial Storytelling and Presentation**

### Document Observations: 
- Each species is being compared with each column in a pairplot
- The Versicolor and Virginica species look closely related, while the Setosa is separated

### **Conclusion:**

Based on the graphs above, the Versicolor & Virginica species are related. They generally have a higher sepal length, petal length, and petal width compared to the Setosa.

The Setosa generally had a higher sepal width than the other two.

The graphs that looks the closest related for each three is when comparing the Sepal Width & Sepal Area.
